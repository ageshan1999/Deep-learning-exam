```python
from datasets import load_dataset

dataset = load_dataset("financial_phrasebank", 'sentences_50agree')

# Select a specific split, for example, "train"
train_dataset = dataset["train"]

# Convert the training dataset to a Pandas DataFrame
df = train_dataset.to_pandas()

# Display the Data
print(df.head())  # You can adjust the number of rows to display as needed
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from datasets import Dataset

# Assuming new_df is your DataFrame
# You may need to adjust the column names and other details based on your actual DataFrame structure

# Step 1: Train-Test Split
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Step 2: Reset Index
train_df = train_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)

# Step 3: Convert to Hugging Face Dataset format
train_dataset = Dataset.from_pandas(train_df)
test_dataset = Dataset.from_pandas(test_df)

# You might want to specify the column names and other details depending on your DataFrame structure.
# For example, if your DataFrame has columns 'text' and 'label', you can use the following:
# train_dataset = Dataset.from_pandas(train_df[['text', 'label']])
# test_dataset = Dataset.from_pandas(test_df[['text', 'label']])

# If your DataFrame has more complex structure, you may need to handle it accordingly.

# Now, train_dataset and test_dataset are in the Hugging Face Dataset format and you can use them for further processing.
```

```python
small_train_dataset = train_dataset.shuffle(seed=42).select(range(16))
small_eval_dataset = test_dataset.shuffle(seed=42).select(range(16))
```

```python
small_train_dataset
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")


def tokenize_function(examples):
  return tokenizer(examples["text"], padding="max_length", truncation=True)


tokenized_small_train_dataset = train_dataset.map(tokenize_function, batched=True)
tokenized_small_eval_dataset = test_dataset.map(tokenize_function, batched=True)
```

```python
tokenized_small_train_dataset.shape
```

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained("bert-base-cased", num_labels=3)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(output_dir="test_trainer")
```

```python
!pip install evaluate --q
```

```python
import numpy as np
import evaluate

metric = evaluate.load("accuracy")
```

```python
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=1,
    per_device_train_batch_size=4,
    evaluation_strategy="epoch",
    logging_dir='./logs',
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_small_train_dataset,
    eval_dataset=tokenized_small_eval_dataset,
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import gradio as gr

# # Load the model and tokenizer
# model_path = 'path/to/your/saved_model_directory'
# model = AutoModelForSequenceClassification.from_pretrained(model_path)
# tokenizer = AutoTokenizer.from_pretrained(model_path)

# Define the predict function
def predict_label(text):
    # Tokenize the input text
    inputs = tokenizer(text, return_tensors="pt", padding="max_length", truncation=True)

    # Move the input tensor to the same device as the model
    inputs = {key: value.to(model.device) for key, value in inputs.items()}

    # Make prediction using the loaded model
    logits = model(**inputs).logits
    predicted_label = logits.argmax().item()

    return predicted_label

# Create a Gradio interface
iface = gr.Interface(fn=predict_label, inputs="text", outputs="text")

# Launch the interface
iface.launch()
```
